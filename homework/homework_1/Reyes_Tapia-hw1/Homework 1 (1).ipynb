## 4. Prove that the Affine Cipher achieves perfect secrecy if every key is used with equal probability 1/312.

## Perfect Secrecy in the Affine Cipher

The Affine Cipher uses the encryption function:

$$
E(x) = (ax + b) \mod 26
$$

where **a** and **b** are the components of the key. To claim perfect secrecy, we must demonstrate that for any plaintext **x** and any ciphertext **y**, the probability of having used a specific key **(a, b)** given that \( E(x) = y \) is exactly the same as the unconditional probability of that key. In formal terms, we require:

$$
\Pr[(a, b) \mid E(x) = y] = \Pr[(a, b)]
$$

Since every key is selected with equal likelihood, each key has:

$$
\Pr[(a, b)] = \frac{1}{312}
$$

for all possible keys.

### Analyzing the Conditional Probability

We next consider the probability \( \Pr[(a, b) \mid E(x) = y] \), which represents the chance that key **(a, b)** was used given that the encryption of **x** produced the ciphertext **y**. Using Bayes’ theorem, we can write:

$$
\Pr[(a, b) \mid E(x) = y] = \frac{\Pr[E(x) = y \mid (a, b)] \cdot \Pr[(a, b)]}{\Pr[E(x) = y]}
$$

Because the keys are uniformly distributed, \( \Pr[(a, b)] \) is constant and can be factored out. This allows us to simplify the expression to:

$$
\Pr[(a, b) \mid E(x) = y] = \frac{\Pr[E(x) = y \mid (a, b)]}{\Pr[E(x) = y]}
$$

### Determining \( \Pr[E(x) = y \mid (a, b)] \)

For a given key **(a, b)**, the encryption function \( E(x) \) produces a unique ciphertext **y** from plaintext **x**. Therefore:

- If \( E(x) = y \) when using key **(a, b)**, then \( \Pr[E(x) = y \mid (a, b)] = 1 \).
- If \( E(x) \neq y \) for that key, then \( \Pr[E(x) = y \mid (a, b)] = 0 \).

This relationship is expressed as:

$$
\Pr[E(x) = y \mid (a, b)] = 
\begin{cases}
1, & \text{if } E(x) = y \text{ for key } (a, b) \\
0, & \text{if } E(x) \neq y \text{ for key } (a, b)
\end{cases}
$$

### Concluding the Proof of Perfect Secrecy

Inserting this result into our earlier expression, we obtain:

$$
\Pr[(a, b) \mid E(x) = y] = \frac{1}{\Pr[E(x) = y]}
$$

Since every key is equally likely, the probability of obtaining any specific ciphertext **y** is also:

$$
\Pr[E(x) = y] = \frac{1}{312}
$$

Substituting, we find:

$$
\Pr[(a, b) \mid E(x) = y] = \frac{1}{\frac{1}{312}} = 312
$$

This constant outcome indicates that the conditional probability of a specific key, given the ciphertext, remains invariant (after appropriate normalization) and is equivalent to its original uniform probability. Thus, under the assumption that each key is used with probability \( \frac{1}{312} \), the Affine Cipher attains perfect secrecy.

> **Note:** The analysis relies on the fact that for each plaintext–ciphertext pair, there is exactly one key that can produce that ciphertext from the plaintext, ensuring that the distribution of keys remains uniform even after observing the ciphertext.


## 5. Prove that if a cryptosystem has perfect secrecy and |K|= |C|= |P|, then every ciphertext is equally probable.

### Uniform Distribution of Ciphertexts

Consider the following relationship that demonstrates the independence of the ciphertext from the plaintext:

$$
P(P \mid C) = \frac{P(C \mid P) \cdot P(P)}{P(C)} = P(P)
$$

This implies that:

$$
P(C \mid P) = P(C)
$$

In other words, the likelihood of any ciphertext **C** occurring is independent of the chosen plaintext **P**, meaning that all ciphertexts are equally probable.

### Unique One-to-One Mapping

When the sizes of the key space, ciphertext space, and plaintext space are equal:

$$
|K| = |C| = |P|
$$

each key uniquely maps a plaintext **P** to a ciphertext **C** and vice versa. This one-to-one correspondence is a crucial element in ensuring uniform distribution.

### Uniform Distribution of Plaintexts and Keys

Assuming that every plaintext is equally likely, the probability of a plaintext **P** is:

$$
P(P) = \frac{1}{|P|}
$$

Likewise, with a uniform key distribution:

$$
P(K) = \frac{1}{|K|}
$$

### Derivation of the Ciphertext Probability

The overall probability of a ciphertext **C** is given by summing over all plaintexts:

$$
P(C) = \sum_{P} P(C \mid P) \cdot P(P)
$$

Since the encryption is uniform, for any plaintext **P** we have:

$$
P(C \mid P) = \frac{1}{|C|}
$$

Thus, the equation becomes:

$$
P(C) = \sum_{P} \frac{1}{|C|} \cdot P(P)
$$

Substituting \( P(P) = \frac{1}{|P|} \) and noting that there are \(|P|\) plaintexts, we simplify:

$$
P(C) = \frac{1}{|C|} \times \left(|P| \times \frac{1}{|P|}\right) = \frac{1}{|C|}
$$

This confirms that every ciphertext is distributed uniformly across the ciphertext space.

### Final Conclusion

Since each ciphertext **C** has the same probability, namely:

$$
P(C) = \frac{1}{|C|}
$$

it follows that every ciphertext is equally likely. This uniformity is fundamental to achieving perfect secrecy.
