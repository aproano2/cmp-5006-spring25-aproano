### Setting up ModSecurity

Although ModeSecurity is no longer built for Windows, a precompiled version found in `https://www.apachehaus.com/cgi-bin/download.plx` would help.

- Check the Apache version: for this, open a Terminal, go to the bin directory folder where XAMPP is installed, and run `httpd -v`.
- The version for this Computer is `2.4.58 (Win64)`. Therefore, ModSecurity should be version 2.9.6, available at Apache House `https://www.apachehaus.com/cgi-bin/download.plx`.
- Extract the following files: 

| File                         | Destination                          |
|-----------------------------|--------------------------------------|
| mod_security2.so            | C:\xampp\apache\modules\             |
| modsecurity.conf-recommended → Rename it to modsecurity.conf | C:\xampp\apache\conf\         |

- Go to the Apache Config log, and add the lines:

```
# Load ModSecurity
LoadModule security2_module modules/mod_security2.so
Include conf/extra/modsecurity.conf
```

- Go to `modsecurity.conf` at  `C:\xampp\apache\conf\extra\modsecurity.conf`.
- Inside this file, find `SecRuleEngine DetectionOnly` and replace to `SecRuleEngine On`.

### Logging Configuration

- ModSecurity writes its logs into the file `C:\xampp\apache\logs\modsec_audit.log`.

- Open `modsecurity.conf` again.

- Look for the line `#SecAuditLog logs/modsec_audit.log`, and replace to `SecAuditLog C:/xampp/apache/logs/modsec_audit.log`.

- Look for the line `SecAuditLogParts ABIJDEFHZ` (if not, add them below the line `SecRuleEngine On`). This line means:

| Letter | Meaning                                  |
|--------|------------------------------------------|
| A      | HTTP request headers                     |
| B      | HTTP request body                        |
| I      | Information about the triggered rule     |
| J      | Cookie information (if any)              |
| D      | Data about the decision process          |
| E      | Internal error information               |
| F      | HTTP response status code                |
| H      | HTTP response headers                    |
| Z      | End of the audit log                     |




- Restart Apache via the XAMPP Control Panel and Test it on the Kali Linux VM.

And we are ready to go

### Installation

#### Clone the DVWA Repository from Github

- `cd /var/www/html`
- `git clone https://github.com/ethicalhack3r/DVWA.git`

#### Configure DVWA Configuration File

- `cd DVWA`
- `cp config/config.inc.php.dist config/config.inc.php`
- `nano config/config.inc.php`

- Make sure the values for `db_user` and `db_password` are correct.

#### Configure the DVWA Database

- Open the DVWA on `localhost/DVWA/setup.php` and create the Database (same as Windows)`

#### Installing `ModSecurity` on `WSL`

- Install the required packages on Ubuntu

- `sudo apt-get update`
- `sudo apt-get install libapache2-mod-security2`

#### Configure `ModSecurity`

- Make sure `ModSecurity is enabled`
-`sudo a2enmod security2`


#### Configure `modsecurity.conf` file.

- Modify the configuration file to activate `ModSecurity`:
- `SecRuleEngine On`

#### Install the `OWASP CRS` rules

- Don't have the rules? It is possible to install them from its official repo:
- `cd /etc/modsecurity`
- `sudo git clone https://github.com/coreruleset/coreruleset.git`

#### Include `Apache` Rules

- Open the `Apache` configuration file for `ModSecurity`:
- `sudo nano /etc/apache2/mods-enabled/security2.conf`

- Make sure to include all `OWASP CRS` in the file:
```IncludeOptional /etc/modsecurity/coreruleset/crs-setup.conf
IncludeOptional /etc/modsecurity/coreruleset/rules/*.conf
```

#### Restart `Apache`

- Once configured, restart `Apache` to apply all the changes:
`sudo systemctl restart apache2`.

#### Basic Logging Configuration

- To enable basic logging for `ModSecurity` in `Apache`, make sure that the `logs` file is correctoy configured at `etc/modsecurity/modsecurity.conf`:

```
SecAuditLog /var/log/apache2/modsec_audit.log
SecAuditLogParts ABIFHZ
```

#### Verifying

- Login to DVWA.
- Verify the lofs using `tail -f /var/log/apache2/modsec_audit.log`.

And we are all set!
